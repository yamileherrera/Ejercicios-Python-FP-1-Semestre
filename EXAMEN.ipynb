solucion pregunta 4 Examen algebra lineal

```
import numpy as np
Matriz = np.arange (9)
np.random.shuffle (Matriz)
Matriz = Matriz.reshape((3,3))
print(Matriz)
```

```
B = np.array ([[1],[3],[2]])
print(B)
```

```
inv = np.linalg.inv(Matriz)
inv
```

```
det= np.linalg.det(Matriz)
det
```

```
adj = det*inv
print(adj)
```

```
cof = adj.T
print(cof)
```
