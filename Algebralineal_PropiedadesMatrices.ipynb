```
A=np.array([[2,-2,2],[2,1,0],[3,-2,2]])
B=np.array([[1,-2,5],[3,3,-1],[0,4,-2]])
```

```
AB = A.dot(B)
AB_inv = np.linalg.inv(AB)
AB_inv
```

```
A_inv = np.linalg.inv(A)
B_inv = np.linalg.inv(B)
BA_inv = B_inv@A_inv 
BA_inv
```
