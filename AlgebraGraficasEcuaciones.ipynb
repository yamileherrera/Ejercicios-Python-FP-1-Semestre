```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```
# Damos los valores que tomara la variable x
x = np.arange (-5,5)
```

```
y_1 = 6-2*x
y_2 = (14 -4*x)/3
```

```
plt.figure()
plt.plot(x,y_1)
plt.plot(x,y_2)
# valores o intervalos para la grafica 
plt.xlim(-1,5)
plt.ylim(-1,5)
#color para los ejes
plt.axvline(x=0, color="grey")
plt.axhline(y=0, color="grey")
```

```
A=np.array([[2,1],[4,3]])
x1 = np.array([2,2])
x = x1.T
```

```
b= A.dot(x)
print(b)
```

```
A_inv = np.linalg.inv(A)
sol = A_inv@b
print(sol)
```
